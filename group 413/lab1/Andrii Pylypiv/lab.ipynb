```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gamma_correction(src, gamma):
    hist_gamma = 1 / gamma

    table = [((i / 255) ** hist_gamma) * 255 for i in range(256)]
    table = np.array(table, np.uint8)

    return cv2.LUT(src, table)
```

```python
def showFilter(img, k):
    plt.imshow(img)
    plt.axis('off')
    plt.title("Left shifk with k = " + str(k))
    plt.show()
```

```python
def showCommon(img):
    plt.imshow(img)
    plt.axis('off')
    plt.title("Default image")
    plt.show()
```

```python
imgContrast = cv2.cvtColor(cv2.imread('contrast.jpg'), cv2.COLOR_BGR2RGB)
imgShadow = cv2.cvtColor(cv2.imread('shadow.jpg'), cv2.COLOR_BGR2RGB)
imgLight = cv2.cvtColor(cv2.imread('light.jpg'), cv2.COLOR_BGR2RGB)
```

```python
showCommon(imgContrast)
shift = gamma_correction(imgContrast, 0.4)
showFilter(shift, 0.4)
```

```python
showCommon(imgShadow)
shift = gamma_correction(imgShadow, 0.4)
showFilter(shift, 0.4)
```

```python
showCommon(imgLight)
shift = gamma_correction(imgLight, 0.4)
showFilter(shift, 0.4)
```
