```python
import torch
from PIL import Image

model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
model.eval()

image = Image.open('/content/drive/MyDrive/photos/lviv1.jpg')
results = model([image])

# Display the detected objects and their confidence scores
results.show()
```
