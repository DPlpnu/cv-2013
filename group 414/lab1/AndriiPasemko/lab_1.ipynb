```python
s = 5
a = 0.5
t = 2
```

```python
import numpy as np

def get_contrasted_image(image):
  height, width, channels = image.shape

  contrasted = np.zeros((height, width, channels), dtype=np.uint8)

  for y in range(height):
      for x in range(width):
          for c in range(channels):
              channel = int(a * (image[y, x, c] + s) - t)
              channel = np.clip(channel, 0, 255)

              contrasted[y, x, c] = channel

  return contrasted.astype(np.uint8)
```

```python
import cv2
import matplotlib.pyplot as plt

def print_contrasted_img(path):
  img = cv2.imread(path)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  img = np.array(img)

  contrasted = get_contrasted_image(img)
  plt.imshow(img)
  plt.figure()
  plt.imshow(contrasted)
```
