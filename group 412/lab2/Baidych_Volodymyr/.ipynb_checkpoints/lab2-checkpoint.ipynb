```python
def my_matcher(img1, img2):
    des1, des2, kp1, kp2, pca_des = sift_pca(img1, img2)
    matches = []
    for i, d1 in enumerate(pca_des[:len(des1)]):
        distances = np.linalg.norm(d1-pca_des[len(des1):], axis=1)
        min_idx = np.argmin(distances)
        min_dist = distances[min_idx]
        if min_dist < 0.8:
            matches.append(cv2.DMatch(i, min_idx, min_dist))
    img = cv2.drawMatches(img1, kp1, img2, kp2, matches[:10], img2, flags=2)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```
